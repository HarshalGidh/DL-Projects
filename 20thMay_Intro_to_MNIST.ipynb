```python
def plot_digits(image_num):
  image=x_train[image_num]
  label= y_train[image_num]
  plt.imshow(image, cmap='gray')
  print(f"Label : {label}")
```

```python
#plot random images of similar digits :
def print_digit_images(digit,num= 10):
  label = np.where(y_train==digit)[0]    # collect all the images of that  particular digit
  #label = label[np.random.randint(0,len(image),num)]
  label = np.random.choice(label,num)   #Randomly pick 10 images of that digit
  #Plot the images :
  plt.figure(figsize=(12,10))
  for i in range(num):
    plt.subplot(1,num,i+1)
    plt.title(f"{num} images of {digit}")
    plt.imshow(x_train[label[i]], cmap='gray')
```

```python
print_digit_images(7,5)
```
